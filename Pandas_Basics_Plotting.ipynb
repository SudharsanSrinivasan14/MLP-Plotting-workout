```python
# Data Slicing ......
```

```python
x = df.iloc[:,:-1]
```

```python
y = df.iloc[:,-1]
```

```python
# Plotting ......
```

```python
plt.plot(x,y,"s",ms='9',mec='y', mfc='g',ls=':',color='r',linewidth='3')
plt.title("Students Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.grid(axis="x",color='b')
plt.show()
```

```python
plt.plot(x,y)
plt.title("Students Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.grid()
plt.show()
```

```python
plt.plot(x,y,ls='--')
plt.title("Students Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.grid(axis="y")
plt.show()
```

```python
plt.plot(x,y,ls='-.')
plt.title("Students Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.grid(axis="x")
plt.show()
```
